Attacking Using Linear Regression Correlation Technique

```python
##The below code implements correlation using Linear Regression Techinque---> Result is low correlation

from scipy.stats import pearsonr
key_guess = []
numtraces = trace_array.shape[0]

numpoints = trace_array.shape[1]
start_point = 400
end_point = 1500

crvs = np.zeros((16, 256, numpoints))
HW_matrix = np.zeros((numtraces, 256), dtype=np.uint8)

printable = []    # for printing the top guessed keys


for subkey in tnrange(0, 16, desc="Attacking Subkey"):
    temp = []    # it holds the [(key_guess1, correlation_value),(key_guess2, correlation_value),..]
    for kguess in tnrange(0, 256, desc="Generating Hamming Weights"):
        # Generate the hamming weight for each possible keybyte
        for trace_no in range(numtraces):
            iv = intermediate(textin_array[trace_no][subkey], kguess)
            HW_matrix[trace_no, kguess] = HW[iv]

        # correlate the trace with the hamming weights
        for point in range(start_point, end_point):
            hw = HW_matrix[:, kguess]
            trc = trace_array[:, point]
            crvs[subkey, kguess, point] = np.abs(linregress(hw, trc).slope)

        temp.append((kguess, np.max(crvs[subkey, kguess])))

    temp.sort(key = lambda x: -x[1])  # sort temp by dom value
    printable.append(temp)  # add the data in list
    df = pd.DataFrame(printable).transpose()

    key_guess.append(crvs[subkey].max(axis=1).argmax())

    clear_output(wait=True)  # clear the previous output
    display(df.head().style.format(format_stat).apply(color_corr_key, axis=0))  # display the current status
```

```python
##The below code implements correlation using Pearson Techinque---> Result is high correlation and correct key guess

import numpy as np
import pandas as pd
from IPython.display import display, clear_output

# AES S-box (256 entries)
AES_SBOX = np.array([
    0x63, 0x7C, 0x77, 0x7B, 0xF2, 0x6B, 0x6F, 0xC5,
    0x30, 0x01, 0x67, 0x2B, 0xFE, 0xD7, 0xAB, 0x76,
    0xCA, 0x82, 0xC9, 0x7D, 0xFA, 0x59, 0x47, 0xF0,
    0xAD, 0xD4, 0xA2, 0xAF, 0x9C, 0xA4, 0x72, 0xC0,
    0xB7, 0xFD, 0x93, 0x26, 0x36, 0x3F, 0xF7, 0xCC,
    0x34, 0xA5, 0xE5, 0xF1, 0x71, 0xD8, 0x31, 0x15,
    0x04, 0xC7, 0x23, 0xC3, 0x18, 0x96, 0x05, 0x9A,
    0x07, 0x12, 0x80, 0xE2, 0xEB, 0x27, 0xB2, 0x75,
    0x09, 0x83, 0x2C, 0x1A, 0x1B, 0x6E, 0x5A, 0xA0,
    0x52, 0x3B, 0xD6, 0xB3, 0x29, 0xE3, 0x2F, 0x84,
    0x53, 0xD1, 0x00, 0xED, 0x20, 0xFC, 0xB1, 0x5B,
    0x6A, 0xCB, 0xBE, 0x39, 0x4A, 0x4C, 0x58, 0xCF,
    0xD0, 0xEF, 0xAA, 0xFB, 0x43, 0x4D, 0x33, 0x85,
    0x45, 0xF9, 0x02, 0x7F, 0x50, 0x3C, 0x9F, 0xA8,
    0x51, 0xA3, 0x40, 0x8F, 0x92, 0x9D, 0x38, 0xF5,
    0xBC, 0xB6, 0xDA, 0x21, 0x10, 0xFF, 0xF3, 0xD2,
    0xCD, 0x0C, 0x13, 0xEC, 0x5F, 0x97, 0x44, 0x17,
    0xC4, 0xA7, 0x7E, 0x3D, 0x64, 0x5D, 0x19, 0x73,
    0x60, 0x81, 0x4F, 0xDC, 0x22, 0x2A, 0x90, 0x88,
    0x46, 0xEE, 0xB8, 0x14, 0xDE, 0x5E, 0x0B, 0xDB,
    0xE0, 0x32, 0x3A, 0x0A, 0x49, 0x06, 0x24, 0x5C,
    0xC2, 0xD3, 0xAC, 0x62, 0x91, 0x95, 0xE4, 0x79,
    0xE7, 0xC8, 0x37, 0x6D, 0x8D, 0xD5, 0x4E, 0xA9,
    0x6C, 0x56, 0xF4, 0xEA, 0x65, 0x7A, 0xAE, 0x08,
    0xBA, 0x78, 0x25, 0x2E, 0x1C, 0xA6, 0xB4, 0xC6,
    0xE8, 0xDD, 0x74, 0x1F, 0x4B, 0xBD, 0x8B, 0x8A,
    0x70, 0x3E, 0xB5, 0x66, 0x48, 0x03, 0xF6, 0x0E,
    0x61, 0x35, 0x57, 0xB9, 0x86, 0xC1, 0x1D, 0x9E,
    0xE1, 0xF8, 0x98, 0x11, 0x69, 0xD9, 0x8E, 0x94,
    0x9B, 0x1E, 0x87, 0xE9, 0xCE, 0x55, 0x28, 0xDF,
    0x8C, 0xA1, 0x89, 0x0D, 0xBF, 0xE6, 0x42, 0x68,
    0x41, 0x99, 0x2D, 0x0F, 0xB0, 0x54, 0xBB, 0x16
], dtype=np.uint8)

def hamming_weight(x):
    return bin(x).count("1")

# Highlight the top correlation in green
def color_corr_key(column):
    values = [float(cell.split(':')[1]) for cell in column]
    max_corr = max(values)
    return [
        'background-color: lightgreen; font-weight: bold' if abs(float(cell.split(':')[1]) - max_corr) < 1e-8 else ''
        for cell in column
    ]

def cpa_aes_top5(traces, plaintexts, start=0, end=None):
    if end is None:
        end = traces.shape[1]

    num_traces, num_points = traces.shape
    trace_window = traces[:, start:end].astype(np.float32)
    trace_mean = np.mean(trace_window, axis=0)
    trace_std = np.std(trace_window, axis=0)

    key_guess = []
    printable = []

    for subkey in range(16):
        temp = []

        for kguess in range(256):
            hw = np.array([
                hamming_weight(AES_SBOX[pt ^ kguess])
                for pt in plaintexts[:, subkey]
            ], dtype=np.float32)

            hw_mean = np.mean(hw)
            hw_std = np.std(hw)
            if hw_std == 0:
                continue

            hw_centered = (hw - hw_mean).reshape(-1, 1)
            cov = np.mean(hw_centered * (trace_window - trace_mean), axis=0)
            denom = hw_std * trace_std

            corr = np.zeros_like(cov)
            mask = denom != 0
            corr[mask] = cov[mask] / denom[mask]

            temp.append((kguess, np.max(np.abs(corr))))

        temp.sort(key=lambda x: -x[1])
        top5 = [f"{v[0]:02X}:{v[1]:.5f}" for v in temp[:5]]

        printable.append(top5)
        key_guess.append(temp[0][0])

        # Display update
        df = pd.DataFrame(printable, columns=[f"Top-{i+1}" for i in range(5)]).transpose()
        clear_output(wait=True)
        display(df.style.apply(color_corr_key, axis=0))

    print("\nRecovered Key:")
    print(" ".join(f"{k:02X}" for k in key_guess))
    return key_guess

cpa_aes_top5(trace_array, textin_array)
```
